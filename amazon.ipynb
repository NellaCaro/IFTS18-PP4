### 1.1 TIPOS DE DATOS + VALORES UNICOS POR COLUMNA

```python
# Crear DataFrame con tipos y valores √∫nicos
tabla_completa = pd.DataFrame({
    'Columna': df_clean.columns,
    'Tipo_Dato': df_clean.dtypes.values,
    'Valores_√önicos': [df_clean[col].nunique() for col in df_clean.columns]
})

# Mostrar en tabla
display(tabla_completa)

print(f"\n RESUMEN: {len(tabla_completa)} columnas analizadas")
```

### 1.2 VALORES NULOS

```python
# Solo columnas que tienen nulos
columnas_con_nulos = df_clean.columns[df_clean.isnull().sum() > 0]

print("CLASIFICACI√ìN DE NULOS POR GRAVEDAD:\n")

# Contadores para el resumen
criticos = 0
altos = 0
medios = 0
leves = 0

for columna in columnas_con_nulos:
    nulos = df_clean[columna].isnull().sum()
    porcentaje = (nulos / len(df_clean)) * 100
    
    # Clasificar por gravedad
    if porcentaje > 50:
        gravedad = "üî¥ CR√çTICO"
        criticos += 1
    elif porcentaje > 10:
        gravedad = "üü° ALTO"
        altos += 1
    elif porcentaje > 1:
        gravedad = "üü† MEDIO" 
        medios += 1
    else:
        gravedad = "üü¢ LEVE"
        leves += 1
    
    print(f"{gravedad} {columna}:")
    print(f"   ‚Ä¢ {nulos} nulos ({porcentaje:.1f}%)")

print(f"\n RESUMEN DE GRAVEDAD:")
print(f"   üî¥ CR√çTICOS: {criticos} columnas (>50% nulos)")
print(f"   üü° ALTOS: {altos} columnas (10-50% nulos)")
print(f"   üü† MEDIOS: {medios} columnas (1-10% nulos)") 
print(f"   üü¢ LEVES: {leves} columnas (<1% nulos)")
print(f"    TOTAL: {len(columnas_con_nulos)} columnas con nulos")
```

### 1.3 DETECCI√ìN DE REDUNDANCIAS

```python
redundancias = []
constantes = []

# 1. Verificar columnas que podr√≠an ser duplicadas
print(" POSIBLES REDUNDANCIAS:")
if 'b2b' in df_clean.columns and 'is_b2b' in df_clean.columns:
    son_iguales = (df_clean['b2b'] == df_clean['is_b2b']).all()
    print(f"   ‚Ä¢ b2b vs is_b2b: ¬øSon iguales? {son_iguales}")
    if son_iguales:
        redundancias.append(('b2b', 'is_b2b', 'Columnas id√©nticas'))

# 2. Buscar columnas constantes (un solo valor)
print("\n COLUMNAS CONSTANTES:")
for columna in df_clean.columns:
    valores_unicos = df_clean[columna].nunique()
    if valores_unicos == 1:
        valor = df_clean[columna].iloc[0]
        print(f"   ‚Ä¢ {columna}: Solo tiene un valor -> {valor}")
        constantes.append((columna, valor))

# 3. Verificar columnas redundantes por significado
print("\n COLUMNAS POTENCIALMENTE INNECESARIAS:")
if 'index' in df_clean.columns:
    print("   ‚Ä¢ index: Redundante con √≠ndice autom√°tico de pandas")
    redundancias.append(('index', '√≠ndice_autom√°tico', 'Duplicado del √≠ndice'))

if 'currency' in df_clean.columns:
    valores_currency = df_clean['currency'].value_counts()
    if len(valores_currency) == 1:
        print(f"   ‚Ä¢ currency: Solo tiene un valor -> {valores_currency.index[0]}")
        constantes.append(('currency', valores_currency.index[0]))

if 'ship_country' in df_clean.columns:
    valores_pais = df_clean['ship_country'].value_counts()
    if len(valores_pais) == 1:
        print(f"   ‚Ä¢ ship_country: Solo tiene un valor -> {valores_pais.index[0]}")
        constantes.append(('ship_country', valores_pais.index[0]))

print(f"\n RESUMEN REDUNDANCIAS:")
print(f"   ‚Ä¢ Columnas redundantes identificadas: {len(redundancias)}")
print(f"   ‚Ä¢ Columnas constantes identificadas: {len(constantes)}")
```

## FASE 2: LIMPIEZA

EXPLICACI√ìN DETALLADA DE CADA PROBLEMA

*** b2b vs is_b2b: Columnas id√©nticas
PROBLEMA: Informaci√≥n duplicada
Ambas columnas contienen EXACTAMENTE la misma informaci√≥n
Ambas tienen 2 valores √∫nicos: True/False
Ambas miden si la venta es Business-to-Business
Ocupan espacio innecesario en la base de datos
SOLUCI√ìN: Mantener solo una (is_b2b es m√°s descriptiva)

*** index: Redundante con √≠ndice autom√°tico
PROBLEMA: Columna innecesaria
Es un duplicado del √≠ndice autom√°tico que pandas ya crea (0, 1, 2, 3...)
No agrega informaci√≥n de negocio
OcUpa memoria sin aportar valor
SOLUCI√ìN: Eliminar - ya tenemos indexaci√≥n autom√°tica

*** currency: Valor constante 'INR'
PROBLEMA: Informaci√≥n redundante
Solo tiene un valor: 'INR' (Rupia India)
Todas las transacciones son en la misma moneda
No ayuda en segmentaci√≥n o filtros
En Tableau, podemos poner 'INR' como etiqueta fija
SOLUCI√ìN: Eliminar - informaci√≥n constante

*** ship_country: Valor constante 'IN'
PROBLEMA: Informaci√≥n redundante
Solo tiene un valor: 'IN' (India)
Todas las ventas son dentro de India
No permite an√°lisis geogr√°fico comparativo
Ocupa espacio sin variabilidad
SOLUCI√ìN: Eliminar - todos los env√≠os son nacionales

*** is_negative_revenue: Valor constante 'False'
PROBLEMA: M√©trica in√∫til
Solo tiene un valor: False
Indica que NINGUNA venta tuvo ingresos negativos
No aporta informaci√≥n para el an√°lisis
SOLUCI√ìN: Eliminar - no hay casos de ingresos negativos

*** order_year: Valor constante '2022'
PROBLEMA: Informaci√≥n redundante
Solo tiene un valor: 2022
Todos los datos son del mismo a√±o
Ya tenemos la columna 'date' para an√°lisis temporal
SOLUCI√ìN: Eliminar - informaci√≥n constante y redundante

*** order_hour: Valor constante '0'
PROBLEMA: Datos incorrectos o sin variabilidad
Solo tiene un valor: 0 (medianoche)
Es improbable que TODAS las ventas sean exactamente a la misma hora
Sugiere problema en la captura de datos
SOLUCI√ìN: Eliminar - datos probablemente err√≥neos

PROBLEMAS CR√çTICOS DE NULOS
üî¥ fulfilled_by: 72.4% nulos
PROBLEMA: Columna casi vac√≠a
72% de los datos faltan
Solo el 28% tiene valor ("Easy Ship")
Demasiados nulos para ser √∫til en an√°lisis
SOLUCI√ìN: Eliminar - demasiada informaci√≥n perdida

üü° unnamed_22: 37.9% nulos
PROBLEMA: Columna misteriosa y con muchos nulos
38% de datos faltantes
No tiene nombre descriptivo
Cuando tiene valor, solo es "False"
SOLUCI√ìN: Eliminar - sin significado + muchos nulos

üü† promotion_ids: 35.5% nulos
PROBLEMA: Muchos nulos pero potencialmente √∫til
35% de datos faltantes
Los nulos significan "sin promoci√≥n"
Podr√≠a ser √∫til para an√°lisis de marketing
SOLUCI√ìN: Mantener pero rellenar nulos con "Sin promoci√≥n"

üü† amount/gross_revenue: 6.3% nulos
PROBLEMA: Nulos en datos financieros cr√≠ticos
6% de √≥rdenes sin monto
Probablemente son √≥rdenes canceladas
SOLUCI√ìN: Investigar y decidir (eliminar o marcar como canceladas)

RESUMEN COMPLETO:
ELIMINAR POR NULOS:
fulfilled_by (72% nulos)
unnamed_22 (38% nulos)

ELIMINAR POR REDUNDANCIA/CONSTANTES:
index, currency, ship_country, is_negative_revenue, order_year, order_hour, b2b

MANTENER Y ARREGLAR:
promotion_ids (rellenar nulos)
amount/gross_revenue (investigar nulos)

### 2.1 ELIMINACI√ìN POR NULOS

```python
# Columnas con muchos nulos (basado en diagn√≥stico FASE 1)
columnas_nulos_criticos = [
    'fulfilled_by',    # üî¥ 72.4% nulos - demasiado perdido
    'unnamed_22'       # üü° 37.9% nulos + sin significado
]

print(" ELIMINANDO COLUMNAS CON MUCHOS NULOS:")
for columna in columnas_nulos_criticos:
    if columna in df_clean.columns:
        nulos = df_clean[columna].isnull().sum()
        print(f"   ‚Ä¢ {columna}: {nulos} nulos ({(nulos/len(df_clean)*100):.1f}%)")

print(f"\n ANTES: {df_clean.shape[1]} columnas")

# Ejecutar eliminaci√≥n
df_clean = df_clean.drop(columns=columnas_nulos_criticos, errors='ignore')

print(f" DESPU√âS: {df_clean.shape[1]} columnas")
print(f" Eliminadas: {len(columnas_nulos_criticos)} columnas con muchos nulos")
```

### 2.2: ELIMINACI√ìN POR REDUNDANCIAS Y CONSTANTES

```python
# Columnas redundantes y constantes
columnas_redundantes = [
    'index', 'currency', 'ship_country', 'is_negative_revenue', 
    'order_year', 'order_hour', 'b2b'
]

print(f" ANTES: {df_clean.shape[1]} columnas")

# Ejecutar eliminaci√≥n
df_clean = df_clean.drop(columns=columnas_redundantes, errors='ignore')

print(f" DESPU√âS: {df_clean.shape[1]} columnas")
print(f" Eliminadas: {len(columnas_redundantes)} columnas redundantes/constantes")
```

### 2.3 MANEJO DE NULOS EN COLUMNAS IMPORTANTES

```python
# 1. Rellenar nulos en promotion_ids
print("1.  RELLENANDO PROMOTION_IDS:")
nulos_promo = df_clean['promotion_ids'].isnull().sum()
print(f"   ‚Ä¢ Nulos antes: {nulos_promo}")

df_clean['promotion_ids'] = df_clean['promotion_ids'].fillna('Sin promoci√≥n')

nulos_promo_despues = df_clean['promotion_ids'].isnull().sum()
print(f"   ‚Ä¢Nulos despu√©s: {nulos_promo_despues}")
print(f"    Nulos arreglados: {nulos_promo}")

# 2. Investigar nulos en amount/gross_revenue
print("\n2.  INVESTIGANDO NULOS EN AMOUNT:")
nulos_amount = df_clean['amount'].isnull().sum()
print(f"   ‚Ä¢ Nulos en amount: {nulos_amount}")

if nulos_amount > 0:
    # Verificar estados de las √≥rdenes sin amount
    ordenes_sin_amount = df_clean[df_clean['amount'].isnull()]
    print(f"   ‚Ä¢ Estados de √≥rdenes sin amount:")
    estados = ordenes_sin_amount['status'].value_counts()
    for estado, cantidad in estados.items():
        print(f"     - {estado}: {cantidad} √≥rdenes")

print(f"\n RESUMEN FASE 2.3:")
print(f"   ‚Ä¢ Promotion_ids: {nulos_promo} nulos ‚Üí 0 nulos")
print(f"   ‚Ä¢ Amount: {nulos_amount} nulos por investigar")
```

### 2.4 MANEJO DE √ìRDENES SIN MONTO + DOCUMENTACI√ìN

```python
# Crear columna para identificar cancelaciones
print("1.  CREANDO COLUMNA DE IDENTIFICACI√ìN:")
df_clean['is_cancelled'] = df_clean['status'] == 'Cancelled'
cancelaciones = df_clean['is_cancelled'].sum()
print(f"   ‚Ä¢ √ìrdenes canceladas identificadas: {cancelaciones}")

# 2. Para las canceladas, establecer amount = 0
print("\n2.  AJUSTANDO MONTOS DE CANCELADAS:")
ordenes_sin_monto = df_clean['amount'].isnull().sum()
print(f"   ‚Ä¢ √ìrdenes sin monto antes: {ordenes_sin_monto}")

# Poner amount = 0 solo para canceladas
df_clean.loc[df_clean['is_cancelled'], 'amount'] = 0
df_clean.loc[df_clean['is_cancelled'], 'gross_revenue'] = 0

ordenes_sin_monto_despues = df_clean['amount'].isnull().sum()
print(f"   ‚Ä¢ √ìrdenes sin monto despu√©s: {ordenes_sin_monto_despues}")
print(f"    Montos ajustados en {cancelaciones} √≥rdenes canceladas")
```

```python
# AN√ÅLISIS DETALLADO DE √ìRDENES ENVIADAS SIN MONTO

ordenes_problematicas = df_clean[(df_clean['amount'].isnull()) & (~df_clean['is_cancelled'])]

print(" HALLAZGOS CR√çTICOS:")

# 1. Verificar si tienen otros datos de venta
print("\n1.  DATOS FINANCIEROS RELACIONADOS:")
print(f"   ‚Ä¢ Tienen order_total: {ordenes_problematicas['order_total'].notna().sum()}")
print(f"   ‚Ä¢ Tienen gross_revenue: {ordenes_problematicas['gross_revenue'].notna().sum()}")
print(f"   ‚Ä¢ Tienen aov_by_order: {ordenes_problematicas['aov_by_order'].notna().sum()}")

# 2. Verificar si son del mismo canal o categor√≠a
print("\n2.  PATRONES COMUNES:")
print(f"   ‚Ä¢ Canal de ventas: {ordenes_problematicas['sales_channel'].value_counts().to_dict()}")
print(f"   ‚Ä¢ Categor√≠as: {ordenes_problematicas['category'].value_counts().to_dict()}")

# 3. Verificar si tienen cantidad (qty)
print(f"\n3.  INFORMACI√ìN DE PRODUCTO:")
print(f"   ‚Ä¢ Tienen cantidad (qty): {ordenes_problematicas['qty'].notna().sum()}")
if ordenes_problematicas['qty'].notna().any():
    print(f"   ‚Ä¢ Cantidades: {ordenes_problematicas['qty'].value_counts().to_dict()}")

# 4. Verificar si son √≥rdenes con promociones especiales
print(f"\n4.  PROMOCIONES:")
print(f"   ‚Ä¢ Tienen promotion_ids: {ordenes_problematicas['promotion_ids'].notna().sum()}")
```

### 2.5 ELIMINACI√ìN DE LAS 229 √ìRDENES PROBLEM√ÅTICAS

```python
# Registrar estado antes de la eliminaci√≥n
filas_antes = df_clean.shape[0]
nulos_antes = df_clean['amount'].isnull().sum()

print(" ESTADO ANTES:")
print(f"   ‚Ä¢ Filas totales: {filas_antes:,}")
print(f"   ‚Ä¢ √ìrdenes sin amount: {nulos_antes}")
print(f"   ‚Ä¢ Porcentaje problem√°tico: {(nulos_antes/filas_antes*100):.2f}%")

# Eliminar √≥rdenes sin amount (las 229 problem√°ticas)
df_clean = df_clean.dropna(subset=['amount'])

# Registrar estado despu√©s
filas_despues = df_clean.shape[0]
nulos_despues = df_clean['amount'].isnull().sum()
```

## FASE 3: OPTIMIZACI√ìN DE TIPOS DE DATOS

```python
# 1. Fecha ‚Üí datetime (CR√çTICO para Tableau)
print("1.  CONVIRTIENDO FECHA:")
print(f"   ‚Ä¢ Antes: {df_clean['date'].dtype}")
df_clean['date'] = pd.to_datetime(df_clean['date'])
print(f"   ‚Ä¢ Despu√©s: {df_clean['date'].dtype}")

# 2. Cantidad ‚Üí entero
print(f"\n2.  CONVIRTIENDO CANTIDAD:")
print(f"   ‚Ä¢ Antes: {df_clean['qty'].dtype}")
df_clean['qty'] = df_clean['qty'].astype('int64')
print(f"   ‚Ä¢ Despu√©s: {df_clean['qty'].dtype}")

# 3. Optimizar categor√≠as principales
print(f"\n3.  OPTIMIZANDO CATEGOR√çAS:")
categorias_optimizar = ['category', 'size', 'status', 'fulfilment']
for col in categorias_optimizar:
    if col in df_clean.columns:
        print(f"   ‚Ä¢ {col}: object ‚Üí category")
        df_clean[col] = df_clean[col].astype('category')
```

## FASE 4: EXPORTACI√ìN FINAL

```python
# Guardar dataset limpio
archivo_final = 'amazon_sales_clean_tableau.csv'
df_clean.to_csv(archivo_final, index=False)

print(f" DATASET EXPORTADO: {archivo_final}")
print(f" RESUMEN FINAL:")
print(f"   ‚Ä¢ Filas: {df_clean.shape[0]:,}")
print(f"   ‚Ä¢ Columnas: {df_clean.shape[1]}")
print(f"   ‚Ä¢ Tama√±o: {df_clean.memory_usage(deep=True).sum() / 1024**2:.1f} MB")
```
