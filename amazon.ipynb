## FASE 1: DIAGNÓSTICO INICIAL

### 1.1 TIPOS DE DATOS + VALORES UNICOS POR COLUMNA

```python
# Crear DataFrame con tipos y valores únicos
tabla_completa = pd.DataFrame({
    'Columna': df_clean.columns,
    'Tipo_Dato': df_clean.dtypes.values,
    'Valores_Únicos': [df_clean[col].nunique() for col in df_clean.columns]
})

# Mostrar en tabla
display(tabla_completa)

print(f"\n RESUMEN: {len(tabla_completa)} columnas analizadas")
```

### 1.2 VALORES NULOS

```python
# Solo columnas que tienen nulos
columnas_con_nulos = df_clean.columns[df_clean.isnull().sum() > 0]

print("CLASIFICACIÓN DE NULOS POR GRAVEDAD:\n")

# Contadores para el resumen
criticos = 0
altos = 0
medios = 0
leves = 0

for columna in columnas_con_nulos:
    nulos = df_clean[columna].isnull().sum()
    porcentaje = (nulos / len(df_clean)) * 100
    
    # Clasificar por gravedad
    if porcentaje > 50:
        gravedad = "🔴 CRÍTICO"
        criticos += 1
    elif porcentaje > 10:
        gravedad = "🟡 ALTO"
        altos += 1
    elif porcentaje > 1:
        gravedad = "🟠 MEDIO" 
        medios += 1
    else:
        gravedad = "🟢 LEVE"
        leves += 1
    
    print(f"{gravedad} {columna}:")
    print(f"   • {nulos} nulos ({porcentaje:.1f}%)")

print(f"\n RESUMEN DE GRAVEDAD:")
print(f"   🔴 CRÍTICOS: {criticos} columnas (>50% nulos)")
print(f"   🟡 ALTOS: {altos} columnas (10-50% nulos)")
print(f"   🟠 MEDIOS: {medios} columnas (1-10% nulos)") 
print(f"   🟢 LEVES: {leves} columnas (<1% nulos)")
print(f"    TOTAL: {len(columnas_con_nulos)} columnas con nulos")
```

### 1.3 DETECCIÓN DE REDUNDANCIAS

```python
redundancias = []
constantes = []

# 1. Verificar columnas que podrían ser duplicadas
print(" POSIBLES REDUNDANCIAS:")
if 'b2b' in df_clean.columns and 'is_b2b' in df_clean.columns:
    son_iguales = (df_clean['b2b'] == df_clean['is_b2b']).all()
    print(f"   • b2b vs is_b2b: ¿Son iguales? {son_iguales}")
    if son_iguales:
        redundancias.append(('b2b', 'is_b2b', 'Columnas idénticas'))

# 2. Buscar columnas constantes (un solo valor)
print("\n COLUMNAS CONSTANTES:")
for columna in df_clean.columns:
    valores_unicos = df_clean[columna].nunique()
    if valores_unicos == 1:
        valor = df_clean[columna].iloc[0]
        print(f"   • {columna}: Solo tiene un valor -> {valor}")
        constantes.append((columna, valor))

# 3. Verificar columnas redundantes por significado
print("\n COLUMNAS POTENCIALMENTE INNECESARIAS:")
if 'index' in df_clean.columns:
    print("   • index: Redundante con índice automático de pandas")
    redundancias.append(('index', 'índice_automático', 'Duplicado del índice'))

if 'currency' in df_clean.columns:
    valores_currency = df_clean['currency'].value_counts()
    if len(valores_currency) == 1:
        print(f"   • currency: Solo tiene un valor -> {valores_currency.index[0]}")
        constantes.append(('currency', valores_currency.index[0]))

if 'ship_country' in df_clean.columns:
    valores_pais = df_clean['ship_country'].value_counts()
    if len(valores_pais) == 1:
        print(f"   • ship_country: Solo tiene un valor -> {valores_pais.index[0]}")
        constantes.append(('ship_country', valores_pais.index[0]))

print(f"\n RESUMEN REDUNDANCIAS:")
print(f"   • Columnas redundantes identificadas: {len(redundancias)}")
print(f"   • Columnas constantes identificadas: {len(constantes)}")
```

## FASE 2: LIMPIEZA

EXPLICACIÓN DETALLADA DE CADA PROBLEMA

*** b2b vs is_b2b: Columnas idénticas
PROBLEMA: Información duplicada
Ambas columnas contienen EXACTAMENTE la misma información
Ambas tienen 2 valores únicos: True/False
Ambas miden si la venta es Business-to-Business
Ocupan espacio innecesario en la base de datos
SOLUCIÓN: Mantener solo una (is_b2b es más descriptiva)

*** index: Redundante con índice automático
PROBLEMA: Columna innecesaria
Es un duplicado del índice automático que pandas ya crea (0, 1, 2, 3...)
No agrega información de negocio
OcUpa memoria sin aportar valor
SOLUCIÓN: Eliminar - ya tenemos indexación automática

*** currency: Valor constante 'INR'
PROBLEMA: Información redundante
Solo tiene un valor: 'INR' (Rupia India)
Todas las transacciones son en la misma moneda
No ayuda en segmentación o filtros
En Tableau, podemos poner 'INR' como etiqueta fija
SOLUCIÓN: Eliminar - información constante

*** ship_country: Valor constante 'IN'
PROBLEMA: Información redundante
Solo tiene un valor: 'IN' (India)
Todas las ventas son dentro de India
No permite análisis geográfico comparativo
Ocupa espacio sin variabilidad
SOLUCIÓN: Eliminar - todos los envíos son nacionales

*** is_negative_revenue: Valor constante 'False'
PROBLEMA: Métrica inútil
Solo tiene un valor: False
Indica que NINGUNA venta tuvo ingresos negativos
No aporta información para el análisis
SOLUCIÓN: Eliminar - no hay casos de ingresos negativos

*** order_year: Valor constante '2022'
PROBLEMA: Información redundante
Solo tiene un valor: 2022
Todos los datos son del mismo año
Ya tenemos la columna 'date' para análisis temporal
SOLUCIÓN: Eliminar - información constante y redundante

*** order_hour: Valor constante '0'
PROBLEMA: Datos incorrectos o sin variabilidad
Solo tiene un valor: 0 (medianoche)
Es improbable que TODAS las ventas sean exactamente a la misma hora
Sugiere problema en la captura de datos
SOLUCIÓN: Eliminar - datos probablemente erróneos

PROBLEMAS CRÍTICOS DE NULOS
🔴 fulfilled_by: 72.4% nulos
PROBLEMA: Columna casi vacía
72% de los datos faltan
Solo el 28% tiene valor ("Easy Ship")
Demasiados nulos para ser útil en análisis
SOLUCIÓN: Eliminar - demasiada información perdida

🟡 unnamed_22: 37.9% nulos
PROBLEMA: Columna misteriosa y con muchos nulos
38% de datos faltantes
No tiene nombre descriptivo
Cuando tiene valor, solo es "False"
SOLUCIÓN: Eliminar - sin significado + muchos nulos

🟠 promotion_ids: 35.5% nulos
PROBLEMA: Muchos nulos pero potencialmente útil
35% de datos faltantes
Los nulos significan "sin promoción"
Podría ser útil para análisis de marketing
SOLUCIÓN: Mantener pero rellenar nulos con "Sin promoción"

🟠 amount/gross_revenue: 6.3% nulos
PROBLEMA: Nulos en datos financieros críticos
6% de órdenes sin monto
Probablemente son órdenes canceladas
SOLUCIÓN: Investigar y decidir (eliminar o marcar como canceladas)

RESUMEN COMPLETO:
ELIMINAR POR NULOS:
fulfilled_by (72% nulos)
unnamed_22 (38% nulos)

ELIMINAR POR REDUNDANCIA/CONSTANTES:
index, currency, ship_country, is_negative_revenue, order_year, order_hour, b2b

promotion_ids (rellenar nulos)
amount/gross_revenue (investigar nulos)

### 2.1 ELIMINACIÓN POR NULOS

```python
# Columnas con muchos nulos (basado en diagnóstico FASE 1)
columnas_nulos_criticos = [
    'fulfilled_by',    # 🔴 72.4% nulos - demasiado perdido
    'unnamed_22'       # 🟡 37.9% nulos + sin significado
]

print(" ELIMINANDO COLUMNAS CON MUCHOS NULOS:")
for columna in columnas_nulos_criticos:
    if columna in df_clean.columns:
        nulos = df_clean[columna].isnull().sum()
        print(f"   • {columna}: {nulos} nulos ({(nulos/len(df_clean)*100):.1f}%)")

print(f"\n ANTES: {df_clean.shape[1]} columnas")

# Ejecutar eliminación
df_clean = df_clean.drop(columns=columnas_nulos_criticos, errors='ignore')

print(f" DESPUÉS: {df_clean.shape[1]} columnas")
print(f" Eliminadas: {len(columnas_nulos_criticos)} columnas con muchos nulos")
```

### 2.2: ELIMINACIÓN POR REDUNDANCIAS Y CONSTANTES

```python
# Columnas redundantes y constantes
columnas_redundantes = [
    'index', 'currency', 'ship_country', 'is_negative_revenue', 
    'order_year', 'order_hour', 'b2b'
]

print(f" ANTES: {df_clean.shape[1]} columnas")

# Ejecutar eliminación
df_clean = df_clean.drop(columns=columnas_redundantes, errors='ignore')

print(f" DESPUÉS: {df_clean.shape[1]} columnas")
print(f" Eliminadas: {len(columnas_redundantes)} columnas redundantes/constantes")
```

### 2.3 MANEJO DE NULOS EN COLUMNAS IMPORTANTES

```python
# 1. Rellenar nulos en promotion_ids
print("1.  RELLENANDO PROMOTION_IDS:")
nulos_promo = df_clean['promotion_ids'].isnull().sum()
print(f"   • Nulos antes: {nulos_promo}")

df_clean['promotion_ids'] = df_clean['promotion_ids'].fillna('Sin promoción')

nulos_promo_despues = df_clean['promotion_ids'].isnull().sum()
print(f"   •Nulos después: {nulos_promo_despues}")
print(f"    Nulos arreglados: {nulos_promo}")

# 2. Investigar nulos en amount/gross_revenue
print("\n2.  INVESTIGANDO NULOS EN AMOUNT:")
nulos_amount = df_clean['amount'].isnull().sum()
print(f"   • Nulos en amount: {nulos_amount}")

if nulos_amount > 0:
    # Verificar estados de las órdenes sin amount
    ordenes_sin_amount = df_clean[df_clean['amount'].isnull()]
    print(f"   • Estados de órdenes sin amount:")
    estados = ordenes_sin_amount['status'].value_counts()
    for estado, cantidad in estados.items():
        print(f"     - {estado}: {cantidad} órdenes")

print(f"\n RESUMEN FASE 2.3:")
print(f"   • Promotion_ids: {nulos_promo} nulos → 0 nulos")
print(f"   • Amount: {nulos_amount} nulos por investigar")
```

### 2.4 MANEJO DE ÓRDENES SIN MONTO + DOCUMENTACIÓN

```python
# Crear columna para identificar cancelaciones
print("1.  CREANDO COLUMNA DE IDENTIFICACIÓN:")
df_clean['is_cancelled'] = df_clean['status'] == 'Cancelled'
cancelaciones = df_clean['is_cancelled'].sum()
print(f"   • Órdenes canceladas identificadas: {cancelaciones}")

# 2. Para las canceladas, establecer amount = 0
print("\n2.  AJUSTANDO MONTOS DE CANCELADAS:")
ordenes_sin_monto = df_clean['amount'].isnull().sum()
print(f"   • Órdenes sin monto antes: {ordenes_sin_monto}")

# Poner amount = 0 solo para canceladas
df_clean.loc[df_clean['is_cancelled'], 'amount'] = 0
df_clean.loc[df_clean['is_cancelled'], 'gross_revenue'] = 0

ordenes_sin_monto_despues = df_clean['amount'].isnull().sum()
print(f"   • Órdenes sin monto después: {ordenes_sin_monto_despues}")
print(f"    Montos ajustados en {cancelaciones} órdenes canceladas")
```

```python
# ANÁLISIS DETALLADO DE ÓRDENES ENVIADAS SIN MONTO

ordenes_problematicas = df_clean[(df_clean['amount'].isnull()) & (~df_clean['is_cancelled'])]

print(" HALLAZGOS CRÍTICOS:")

# 1. Verificar si tienen otros datos de venta
print("\n1.  DATOS FINANCIEROS RELACIONADOS:")
print(f"   • Tienen order_total: {ordenes_problematicas['order_total'].notna().sum()}")
print(f"   • Tienen gross_revenue: {ordenes_problematicas['gross_revenue'].notna().sum()}")
print(f"   • Tienen aov_by_order: {ordenes_problematicas['aov_by_order'].notna().sum()}")

# 2. Verificar si son del mismo canal o categoría
print("\n2.  PATRONES COMUNES:")
print(f"   • Canal de ventas: {ordenes_problematicas['sales_channel'].value_counts().to_dict()}")
print(f"   • Categorías: {ordenes_problematicas['category'].value_counts().to_dict()}")

# 3. Verificar si tienen cantidad (qty)
print(f"\n3.  INFORMACIÓN DE PRODUCTO:")
print(f"   • Tienen cantidad (qty): {ordenes_problematicas['qty'].notna().sum()}")
if ordenes_problematicas['qty'].notna().any():
    print(f"   • Cantidades: {ordenes_problematicas['qty'].value_counts().to_dict()}")

# 4. Verificar si son órdenes con promociones especiales
print(f"\n4.  PROMOCIONES:")
print(f"   • Tienen promotion_ids: {ordenes_problematicas['promotion_ids'].notna().sum()}")
```

### 2.5 ELIMINACIÓN DE LAS 229 ÓRDENES PROBLEMÁTICAS

```python
# Registrar estado antes de la eliminación
filas_antes = df_clean.shape[0]
nulos_antes = df_clean['amount'].isnull().sum()

print(" ESTADO ANTES:")
print(f"   • Filas totales: {filas_antes:,}")
print(f"   • Órdenes sin amount: {nulos_antes}")
print(f"   • Porcentaje problemático: {(nulos_antes/filas_antes*100):.2f}%")

# Eliminar órdenes sin amount (las 229 problemáticas)
df_clean = df_clean.dropna(subset=['amount'])

# Registrar estado después
filas_despues = df_clean.shape[0]
nulos_despues = df_clean['amount'].isnull().sum()
```

## FASE 3: OPTIMIZACIÓN DE TIPOS DE DATOS

```python
# 1. Fecha → datetime (CRÍTICO para Tableau)
print("1.  CONVIRTIENDO FECHA:")
print(f"   • Antes: {df_clean['date'].dtype}")
df_clean['date'] = pd.to_datetime(df_clean['date'])
print(f"   • Después: {df_clean['date'].dtype}")

# 2. Cantidad → entero
print(f"\n2.  CONVIRTIENDO CANTIDAD:")
print(f"   • Antes: {df_clean['qty'].dtype}")
df_clean['qty'] = df_clean['qty'].astype('int64')
print(f"   • Después: {df_clean['qty'].dtype}")

# 3. Optimizar categorías principales
print(f"\n3.  OPTIMIZANDO CATEGORÍAS:")
categorias_optimizar = ['category', 'size', 'status', 'fulfilment']
for col in categorias_optimizar:
    if col in df_clean.columns:
        print(f"   • {col}: object → category")
        df_clean[col] = df_clean[col].astype('category')
```

## FASE 4: EXPORTACIÓN FINAL

```python
# Guardar dataset limpio
archivo_final = 'amazon_sales_clean_tableau.csv'
df_clean.to_csv(archivo_final, index=False)

print(f" DATASET EXPORTADO: {archivo_final}")
print(f" RESUMEN FINAL:")
print(f"   • Filas: {df_clean.shape[0]:,}")
print(f"   • Columnas: {df_clean.shape[1]}")
print(f"   • Tamaño: {df_clean.memory_usage(deep=True).sum() / 1024**2:.1f} MB")
```
